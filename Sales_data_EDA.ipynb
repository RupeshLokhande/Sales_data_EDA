# Gender_count

```python
df.columns
```

```python
a = sns.countplot(x = "Gender",data = df) 
sns.set(rc={'figure.figsize':(5,5)})
plt.title('Total count by mel and female')
for bars in a.containers:
    a.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index= False)['Amount'].sum().sort_values(by ='Amount',ascending = False)
```

```python
sales_gen = df.groupby(['Gender'],as_index= False)['Amount'].sum().sort_values(by ='Amount',ascending = False)
sns.barplot(x = 'Gender',y = 'Amount',data = sales_gen)
```

# Age_count

```python
df.columns
```

```python
b = sns.countplot(data = df,x = "Age Group",hue = 'Gender')
for bars in b.containers:
    b.bar_label(bars)
```

# Total_amount vs age group

```python
sales_age = df.groupby(['Age Group'],as_index= False)['Amount'].sum().sort_values(by='Amount',ascending = False)
sns.barplot(x = 'Age Group',y = 'Amount',data = sales_age)
```

# Total numbers of orders in statevise

```python
sales_state = df.groupby(['State'],as_index = False)['Orders'].sum().sort_values(by='Orders',ascending=False)
sns.barplot(y='State',x='Orders',data=sales_state)
sns.set(rc={'figure.figsize':(15,4)})
```

# Marital status

```python
a = sns.countplot(data = df, x= 'Marital_Status')
sns.set(rc={'figure.figsize':(15,4)})
for bar in a.containers:
    a.bar_label(bar)
```

# Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
b = sns.countplot(data = df, x='Occupation')
for bar in b.containers:
    b.bar_label(bar)
```

# Product_category

```python
c = sns.countplot(data = df, y='Product_Category')
sns.set(rc={'figure.figsize':(15,4)})
for bar in c.containers:
    c.bar_label(bar)
```

```python
product_sales_amount = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by ='Amount',ascending=False)
sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(product_sales_amount,y='Product_Category',x='Amount')
```

```python
top_saling_product = df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,6)})
sns.barplot(data=top_saling_product,x='Product_ID',y='Orders')
```

# Conclusion:

In this EDA we can see most of the buyers are female and purchasing power of the females are greater than men.
Most of the buyers age of age group between 26-35 years and Most of the orders come from Uttar Pradesh, Maharashtra,
and Karnataka. Most of the customers are married and work in the IT, aviation, and healthcare industries.
Most sold products are from food, footwear and electronics category.
